```python
! pip install openpyxl
from openpyxl import Workbook
from datetime import datetime

# 現在日付を取得
today = datetime.now()

# ファイル名用（YYYYMMDD）
file_date = today.strftime("%Y%m%d")

# 表示用（YYYY/MM/DD）
display_date = today.strftime("%Y/%m/%d")

# ファイル名作成
file_name = f"請求書_{file_date}.xlsx"

# Workbook作成
wb = Workbook()
ws = wb.active
ws.title = "請求書"

# 固定データ入力
ws["B2"] = "請求書"
ws["B3"] = " "
ws["B4"] = "株式会社ABC"
ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws["B7"] = "担当者名:鈴木一郎 様"

ws["F4"] = "No."
ws["G4"] = "0001"

# 日付
ws["F5"] = "日付"
ws["G5"] = display_date


print(f"{file_name} を作成しました。")

worksheet = wb.active


# 表のヘッダー行を書き込む

header = ['商品名', '数量', '単価', '金額']
start_row = 9   # B9から開始
start_col = 2   # B列

for i, value in enumerate(header):
    ws.cell(row=start_row, column=start_col + i, value=value)

# 複数行のデータをリストdataに格納
data = [
    ['商品A', 2, 10000, 20000],
    ['商品B', 1, 15000, 15000]
]

    
# 10行目のB列から書き込む
start_row = 10   # 開始行
start_col = 2   # B列

for r, row_data in enumerate(data):
    for c, value in enumerate(row_data):
        ws.cell(row=start_row + r, column=start_col + c, value=value)


# 集計処理 

data_start_row = 10
data_end_row = data_start_row + len(data) - 1

amount_col = start_col + 3  # 金額列（E列）

subtotal_row = data_end_row + 2
tax_row = subtotal_row + 1
total_row = tax_row + 1

# 小計
ws.cell(row=subtotal_row, column=amount_col - 1, value="小計")
ws.cell(
    row=subtotal_row,
    column=amount_col,
    value=f"=SUM({chr(64+amount_col)}{data_start_row}:{chr(64+amount_col)}{data_end_row})"
)

# 消費税（10%）
ws.cell(row=tax_row, column=amount_col - 1, value="消費税（10%）")
ws.cell(
    row=tax_row,
    column=amount_col,
    value=f"={chr(64+amount_col)}{subtotal_row}*0.1"
)

# 合計
ws.cell(row=total_row, column=amount_col - 1, value="合計")
ws.cell(
    row=total_row,
    column=amount_col,
    value=f"={chr(64+amount_col)}{subtotal_row}+{chr(64+amount_col)}{tax_row}"
)

# ファイル保存
wb.save(file_name)
```
