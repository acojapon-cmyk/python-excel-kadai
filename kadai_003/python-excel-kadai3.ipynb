```python
import pandas as pd

# 処理対象のファイル名のリストを格納
file_list = ['2022_年間売上表.xlsx', '2023_年間売上表.xlsx']

# 結合後のデータを格納するための空のデータフレームを作成
df = pd.DataFrame()

# ファイルを一つずつ処理する
for file_name in file_list:
    # Excelファイルを読み込む
    temp_df = pd.read_excel(file_name)

    # データフレームの列名を指定
    temp_df.columns = ['売上年', '商品', '金額（千円）']
    
    # concat()でデータフレームを連結
    df = pd.concat([df, temp_df], ignore_index=True)

result = (
    df.groupby(['商品', '売上年'], as_index=False)['金額（千円）']
      .sum()
      .sort_values(by=['商品', '売上年'])
)

cols = ['商品', '売上年', '金額（千円）'] 
df = df[cols]

# Excelファイルを作成する
writer = pd.ExcelWriter('売上集計表.xlsx')

# resultを書き出す
result.to_excel(writer, sheet_name='Sheet1', index=False)

# Excelファイルを保存する
writer.close()

from openpyxl import load_workbook
from openpyxl.styles import PatternFill

# Excelを開く
wb = load_workbook('売上集計表.xlsx')
ws = wb['Sheet1']

# 背景色（薄いグレー）
fill = PatternFill(fill_type='solid', start_color='F2F2F2')

# 1行目（ヘッダー）に適用
for cell in ws[1]:
    cell.fill = fill

# 保存
wb.save('売上集計表.xlsx')
```
