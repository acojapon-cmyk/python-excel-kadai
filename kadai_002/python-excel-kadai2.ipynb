```python
import pandas as pd

# データを定義する
df = pd.DataFrame({
    '日付': [20230517, 20230518, 20230519, 20230520, 20230521],
    '社員名': ['山田', '佐藤', '鈴木', '田中', '高橋'],
    '売上': [100, 200, 150, 300, 250],
    '部門': ['メーカー', '代理店', 'メーカー', '商社', '代理店']
})

# ② 日付変換
df['日付'] = pd.to_datetime(df['日付'], format='%Y%m%d').dt.strftime('%Y-%m-%d')

df = pd.DataFrame(df)

# 「平均売上」列を作成し、列を追加する
df['平均売上'] = df['売上'].mean()

# 優先レベルで分岐して業績ランクを求める関数「prioritize」を定義
def performance(level):
  result = ''
  if level >= df['平均売上'].iloc[0] + 50:
    result = 'A'
  elif level >= df['平均売上'].iloc[0]:
    result = 'B'
  else:
    result = 'C'
  return result

# 「業績ランク」列を作成し、関数「prioritize」を適用して値を設定
df['業績ランク'] = df['売上'].apply(performance)

# Excelファイルを作成する
writer = pd.ExcelWriter('業績.xlsx')

# DataFrameオブジェクトをExcelファイルに書き込む
df.to_excel(writer, sheet_name='Sheet1', index=False)

# Excelファイルを保存する
writer.close()
```
